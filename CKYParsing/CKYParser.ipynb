# Exercise 1 - CKY Recognizer
(6 points)

Finish the `CKYRecognizer` class given below. It takes a grammar (the grammar does not need to be checked for its structure as all rules already fit to the necessary CNF described in the slides), and a sentence for which it generates a table comprising the non-terminal symbols that could be used to generate the sentence (as described in the slides).

The two methods are as follows:
- `Constructor`
  - takes a list of non-terminal to non-terminal rules in the CNF (i.e., each rule is a String pointing to an array of String pairs) as a map (the key is the left side of the rule, the value are the possible right sides)
  - takes a list of lexical rules (i.e., non-terminals pointing to a list of terminals they could be replaced with) as a map (the key is the non-terminal on the left side while the value are the possible tokens on the right side)
- `getParseTable`
  - takes a sentence as a single string
    - the tokens in the given sentence are lowercased
    - the tokens are separated by whitespaces
    - the sentence does not contain any punctuation
  - returns the parsing table as a serialized string with
    - each table cell should be enclosed with `[` and `]`
    - an empty cell should be serialized as `[]`
    - the non-terminals in the cells should be
      - seperated by a comma and a whitespace (i.e., `, `) and
      - sorted alphabetically (i.e., `[A, B]` but not `[B, A]`)
  - in case a terminal symbol is unknown, `null` should be returned

#### Example

Imagine a very simple grammar with the non-terminals $A, B, C, S$ and the terminals $a, b$. $S$ is the start symbol and we have the following rules:

For the sentence `"a b"`, the `getParseTable` method should return the following table:
```
[][A][S]
[][][B, C]
[][][]

```
Please note that
* each line of the table ends with a `\n` character (including the last line)
* the first coloumn and the last line are printed although they are typically empty
* you don't have to make sure whether the table contains a complete parsing tree, i.e., whether the sentence can be parsed using the given grammatic. In this first exercise, you only have to apply the CKY recognizer algorithm.

````java
// package ckyAlgorithm;

import java.util.ArrayList;
import java.util.Collections;
import java.util.HashMap;
import java.util.List;
import java.util.Map;

//YOUR CODE HERE

public class CKYRecognizer {

	// YOUR CODE HERE
	Map<String, String[]> lexicon;
	Map<String, String[][]> grammar;
	int[][] table;
	int count = 0;
	/*
	 * This map is used to store indexes that correspond to a cell of table and
	 * stores the list of String
	 */
	Map<Integer, List<String>> tableValues;

	/*
	 * This map stores each combination of NonTerminal as a key.
	 */
	Map<String, List<String>> newGrammar;

	/**
	 * Constructor.
	 * 
	 * @param grammar
	 *            the non-terminal to non-terminal rules
	 * @param lexicon
	 *            the non-terminal to terminal rules
	 */
	public CKYRecognizer(Map<String, String[][]> grammar, Map<String, String[]> lexicon) {
		// YOUR CODE HERE
		this.grammar = grammar;
		this.lexicon = lexicon;
		tableValues = new HashMap<Integer, List<String>>();
		newGrammar = new HashMap<String, List<String>>();
	}

	/*
	 * search for non terminals for the words from lexicon
	 * 
	 */
	public List<String> searchLexicon(String wordArray) {
		List<String> lexList = new ArrayList<>();
		for (String entry : lexicon.keySet()) {
			if (!lexList.contains(entry)) {
				for (String word : lexicon.get(entry)) {
					if (wordArray.equalsIgnoreCase(word))
						lexList.add(entry);
				}
			}
		}
		return lexList;
	}

	/*
	 * search lexicons for all words in a sentence
	 */
	public boolean fillDiagonal(String[] words) {
		table = new int[words.length][words.length];
		for (int i = 0; i < words.length; i++) {
			List<String> nonTermList = searchLexicon(words[i]);
			if(nonTermList.size()==0)
				return false;
			Collections.sort(nonTermList);
			table[i][i] = ++count;
			tableValues.put(count, nonTermList);
		}
		return true;
	}

	/*
	 * change the way grammar exist
	 */

	public void createNewGrammar() {
		for (String key : grammar.keySet()) {
			String[][] str = grammar.get(key);
			for (int i = 0; i < str.length; i++) {
				String temp = "";
				for (int j = 0; j < 2; j++) {
					temp += str[i][j] + " ";
				}
				temp = temp.trim();
				if (newGrammar.containsKey(temp)) {
					if (!newGrammar.get(temp).contains(key))
						newGrammar.get(temp).add(key);
				} else {
					List<String> list = new ArrayList<>();
					list.add(key);
					newGrammar.put(temp, list);
				}
			}
		}
	}

	/*
	 * search the combination from grammar
	 */
	public List<String> searchCombinations(List<String> s1, List<String> s2) {
		String temp = "";
		List<String> finalList = new ArrayList<>();
		for (int i = 0; i < s1.size(); i++) {
			for (int j = 0; j < s2.size(); j++) {
				temp = s1.get(i) + " " + s2.get(j);
				if (newGrammar.containsKey(temp)){
					for(String gram: newGrammar.get(temp)){
						if(!finalList.contains(gram))
							finalList.add(gram);
					}
				}
			}
		}
		Collections.sort(finalList);
		return finalList;
	}

	/*
	 * Follow bottom up approach now
	 */
	public void fillTable(int length) {
		for (int j = 1; j < length; j++) {
			for (int i = j - 1; i >= 0; i--) {
				for (int k = i; k < j; k++) {
					if(table[i][k]>0 && table[k+1][j]>0){
					List<String> cell = searchCombinations(tableValues.get(table[i][k]),
							tableValues.get(table[k + 1][j]));
					if (cell.size() > 0) {
						if(table[i][j]==0){
						table[i][j] = ++count;
						tableValues.put(count, cell);}
						else{
							int tempCount=table[i][j];
							List<String> tempCell = tableValues.get(tempCount);
							for(String cellToken: cell){
								if(!tempCell.contains(cellToken))
									tempCell.add(cellToken);
							}
						}
					}}
				}
			}
		}
	}

	String[][] row;

	public String displayTable(int length) {
		String check="";
		row = new String[length + 1][length + 1];
		
		for (int i = 0; i < length; i++) {
			row[i][0]="[]";
			check+="[]";
			for (int j = 0; j < length; j++) {
				if (table[i][j] > 0) {
					for (String nonTerm : tableValues.get(table[i][j])) {
						if (row[i][j + 1] == null)
							row[i][j + 1] = "["+nonTerm + ", ";
						else
							row[i][j + 1] += nonTerm + ", ";
					}
					row[i][j + 1] = row[i][j + 1].substring(0, row[i][j + 1].length() - 2)+"]";
//					System.out.print("[" + row[i][j + 1] + "]");
					check+=row[i][j + 1];
				}
				else {
					row[i][j+1]="[]";
					check+="[]";
//					System.out.print("[]");
				}

			}
			check+="\n";
//			System.out.println();
		}
		for(int i=0;i<length+1;i++){
			
			check+="[]";
		}
//			System.out.print("[]");
//		System.out.println();
		return check+"\n";
	}

	/**
	 * Takes the sentence and returns the generated table serialized as a single
	 * String,
	 * 
	 * @param sentence
	 *            the sentence that should be parsed
	 * @return the parsing table serialized as a single String.
	 */
	public String getParseTable(String sentence) {
		String parseTable = "";
		// YOUR CODE HERE
		String[] tokens = sentence.split(" ");
		createNewGrammar();
		boolean flag = fillDiagonal(tokens);
		if(flag == false)
			return null;
		fillTable(tokens.length);
		parseTable=displayTable(tokens.length);
		return parseTable;
	}

}

// This line should make sure that compile errors are directly identified when
// executing this cell
// (the line itself does not produce any meaningful result)
new CKYRecognizer(new HashMap<>(), new HashMap<>());
System.out.println("compiled");
````
